```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from pathlib import Path

font = {'size'   : 12}

matplotlib.rc('font', **font)
plt.rcParams['text.usetex'] = True

img_path = Path("img").absolute()
```

```python
gensim_1 = pd.read_csv("gensim_coherence_first.csv", names=['c_v'])
gensim_2 = pd.read_csv("gensim_coherence_second.csv", names=['c_v'])
gensim_3 = pd.read_csv("gensim_coherence_third.csv", names=['c_v'])
gensim_4 = pd.read_csv("gensim_coherence_fourth.csv", names=['c_v'])
tomoto_1 = pd.read_csv("tomoto_coherence_first.csv", names=['c_v'])
tomoto_2 = pd.read_csv("tomoto_coherence_second.csv", names=['c_v'])
tomoto_3 = pd.read_csv("tomoto_coherence_third.csv", names=['c_v'])
tomoto_4 = pd.read_csv("tomoto_coherence_fourth.csv", names=['c_v'])

gensim = [gensim_1, gensim_2, gensim_3, gensim_4]
tomoto = [tomoto_1, tomoto_2, tomoto_3, tomoto_4]
```

```python
topic_range = range(10, 101)
fig, ax = plt.subplots()

plt.xlabel("Number of topics $K$")
plt.ylabel("Topic coherence $C_v$")

labels = ["2010-2015 Con-LibDem ", "2015-2017 Con", "2017-2019 Con", "2019-present Con"]
for g, t, c, l in zip(gensim, tomoto, matplotlib.colors.TABLEAU_COLORS, labels):
    g['topics'] = topic_range
    t['topics'] = topic_range

    ax.plot(topic_range, t.c_v, c=c, label=l)
    ax.plot(topic_range, g.c_v, c=c, linestyle='dashed')

ax.legend()
plt.savefig(img_path.joinpath("topic_coherence.pdf"))
```
