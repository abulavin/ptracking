```python
from ptracking.database import Fetcher
from ptracking.topic import tomoto_topics
from tomotopy.coherence import Coherence
import tomotopy as tp
from matplotlib import pyplot as plt
import numpy as np

min_topics = 10
max_topics = 101
step_size = 1
```

```python
gvmt_period = 'first'

res = Fetcher.select_columns("processed_content", gvmt_period=gvmt_period)['processed_content']

coherence_values_1t = list()

for num_topics in range(min_topics, max_topics, step_size):
    print('Trying {} topics'.format(num_topics))
    petition_topics, lda_model = tomoto_topics(n=num_topics, iterations=30, tw=tp.TermWeight.IDF, gvmt_period=gvmt_period)
    cm = Coherence(corpus=lda_model, coherence='c_v')
    coherence_values_1t.append(cm.get_score())
```

```python
import numpy as np
np.savetxt('tomoto_coherence_first.csv', coherence_values_1t, delimiter=', ')
```

```python
fig = plt.figure(figsize=(15,10))
plt.plot(range(min_topics, max_topics, step_size), coherence_values_1t)
plt.ylabel('Coherence Score')
plt.xlabel('Number of Topics')
plt.xticks(range(min_topics, max_topics, step_size))
plt.show()
```
