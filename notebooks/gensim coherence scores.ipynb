```python
from ptracking.database import Fetcher

from ptracking.topic import topics
from gensim.models.coherencemodel import CoherenceModel

from matplotlib import pyplot as plt
import numpy as np

min_topics = 10
max_topics = 101
step_size = 1
```

```python
gvmt_period = 'first'

res = Fetcher.select_columns("processed_content", gvmt_period=gvmt_period)['processed_content']

coherence_values_1g = list()

for num_topics in range(min_topics, max_topics, step_size):
    print('Trying {} topics'.format(num_topics))
    petition_topics, lda_model = topics(num_topics=num_topics, gvmt_period=gvmt_period)
    cm = CoherenceModel(model=lda_model, texts=res.tolist())
    coherence_values_1g.append(cm.get_coherence())
```

```python
import numpy as np
np.savetxt('gensim_coherence_first.csv', coherence_values_1g, delimiter=', ')
```

```python
fig = plt.figure(figsize=(15,10))
plt.plot(range(min_topics, max_topics, step_size), coherence_values_1g)
plt.ylabel('Coherence Score')
plt.xlabel('Number of Topics')
plt.xticks(range(min_topics, max_topics, step_size))
plt.show()
```
