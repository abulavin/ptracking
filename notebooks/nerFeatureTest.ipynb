```python
from ptracking.sentiment.corenlp import ner
from ptracking.predict.evaluator import Dataset
import matplotlib.pyplot as plt

ner_df = ner()
data = Dataset().prepare(columns=["created_at"]) # can select other columns here too, probably won't need to
joined = data.join(ner_df)
plt.scatter(joined['CITY'] + joined['MONEY'], joined.signatures)
plt.show()
```
